```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df = df.dropna()
```

```python
x = df.drop(['ID','Y'], axis=1)
y = df['Y']
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
svc_classifier = SVC()
svc_classifier.fit(x_train, y_train)
```

```python
svm_predict = svc_classifier.predict(x_test)
```

```python
svm_acuracy = accuracy_score(y_test, svm_predict)
print("Acurácia do Modelo SVM:", svm_acuracy)
```

```python
#matriz de confusão do svm
cm = confusion_matrix(y_test, svm_predict)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Matriz de Confusão do Modelo SVM')
plt.show()
```

```python
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
random_forest.fit(x_train, y_train)
```

```python
random_forest_pred  = random_forest.predict(x_test)
```

```python
random_forest_accuracy = accuracy_score(y_test, random_forest_pred)
print("Acurácia do Modelo Random Forest:", random_forest_accuracy)
```

```python
cm = confusion_matrix(y_test, random_forest_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Matriz de Confusão do Modelo Random Forest')
```
